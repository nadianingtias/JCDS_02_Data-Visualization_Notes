## Fungsi groupby in Panda DataFrames

```python
df3 = pd.DataFrame([
    {'nama' : 'Andi', 'nilai' : 90, 'biologi':90},
    {'nama' : 'Andi', 'nilai' : 95, 'biologi':95},
    {'nama' : 'Andi', 'nilai' : 92, 'biologi':97},
    {'nama' : 'Andi', 'nilai' : 90, 'biologi':91},
    {'nama' : 'Budi', 'nilai' : 91, 'biologi':92},
    {'nama' : 'Budi', 'nilai' : 94, 'biologi':94},
    {'nama' : 'Budi', 'nilai' : 92, 'biologi':90},
    {'nama' : 'Budi', 'nilai' : 90, 'biologi':80},
])
df3grup = (df3.groupby('nama'))
print(df3grup)
print(df3grup.groups.keys())
print("minimal")
print(df3grup.min())
print('rata rata')
print(df3grup.mean())
print(df3grup.get_group('Andi'))
```

```python
df3grup.mean()[['biologi', 'nilai']]
```

## CONCAT (stacking)

```python
df4 = pd.DataFrame([
    {'nama' : 'Andi', 'nilai' : 90, 'biologi':90},
    {'nama' : 'Andi', 'nilai' : 95, 'biologi':95},
    {'nama' : 'Andi', 'nilai' : 92, 'biologi':97}
])
df5 = pd.DataFrame([
    {'nama' : 'Budi', 'nilai' : 94, 'biologi':94},
    {'nama' : 'Budi', 'nilai' : 92, 'biologi':90},
    {'nama' : 'Budi', 'nilai' : 90, 'biologi':80},
])
dfX = pd.concat([df4, df5])
print(dfX)
dfX.loc[0]

dfXX = pd.concat([df4, df5], ignore_index=True, axis=1,
                join="outer")
dfXX
```

## MERGE data in pandas
### merge = join

```python
df6 = pd.DataFrame([
    {'nama' : 'Andi', 'fis' : 90, 'biologi':90},
    {'nama' : 'Budi', 'fis' : 95, 'biologi':95},
    {'nama' : 'Caca', 'fis' : 92, 'biologi':97}
])
df7 = pd.DataFrame([
    {'nama' : 'Andi', 'mat' : 94, 'kim':94},
    {'nama' : 'Budi', 'mat' : 92, 'kim':90},
    {'nama' : 'Caca', 'mat' : 90, 'kim':80},
    {'nama' : 'Deni', 'mat' : 90, 'kim':80},
])
```

```python
# INNER JOIN- default - mendapatkan irisan dari kedua dataframe
dfXXX = pd.merge(df6,df7, on='nama')
dfXXX
```

```python
# outer = mendapatkan semua unique NAMA
dfXXX = pd.merge(df6,df7, on='nama', how='outer')
dfXXX
```
