```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'x' : np.arange(10),
    'y' : np.arange(10),
    'z' : [ "wanita", "pria", "wanita", "pria", "pria", "wanita", "wanita", "pria", "pria", "wanita"]
})

# %seaborn inline
sns.set_style('whitegrid')
sns.relplot(data=df, x='x', y='y', kind="line"
           ,hue='z', style='z', size='z')
plt.show()
# sns.relplot(x="x", y="y", sort=False, kind="line");
```

```python
plt.show()
sns.relplot(data=df, x='x', y='y'
           ,hue='z', style='z', size='z')
```

```python
sns.barplot(data=df, x = 'x', y='y', hue='z')
plt.show()
```

```python
flight = sns.load_dataset('flights')
yearpassengers = flight.groupby('year').sum()
# flig
yearpassengers
```

```python
sns.barplot(data=yearpassengers, x = yearpassengers.index, y='passengers', orient='v')
plt.show()
```

```python
pivotFlight = flight.pivot('month', 'year', 'passengers')
pivotFlight
# pivotFlight[]
```

```python
sums = pivotFlight.sum()
# print(sums)
sums
```

```python
fig = plt.figure(figsize=(10,10))

sns.heatmap(pivotFlight, annot=True, fmt='d')
plt.show()
```
