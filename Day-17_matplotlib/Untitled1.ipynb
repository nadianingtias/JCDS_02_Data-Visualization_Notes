```python
# menyelesaikan persoalan SPLDV
# x+y = 7
# 3x+y=15
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,7)
y = 15- x*3

x2 = np.arange(0,7)
y2 = 7-x2

plt.plot(x,y)
plt.plot(x2,y2)
plt.axhline(color='k', linewidth=0.3)
plt.axvline(color='k', linewidth=0.3)

matSoal = np.array([[1,1],[3,1]])
matDiket = np.array([7,15])
matJawab = np.linalg.solve(matSoal, matDiket)
print(matJawab)
plt.plot(matJawab[0], matJawab[1], 'ro')
plt.title("Grafik Penyelesaian")
plt.annotate(
         'Titik\npotong '+ str(matJawab)+' ', xy=(matJawab), xytext=(4,4)
)

plt.show()
```
