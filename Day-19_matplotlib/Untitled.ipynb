```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
fig = plt.figure(figsize=(10,10))
p = plt.subplot(111, projection='3d')

x = np.arange(10)
print(x)

p.plot_wireframe(x,(x**2+65),(x**7), facecolor="lightsalmon", alpha=0.2)
p.scatter(x,(x**2+65),(x**7), marker="*", color='g', s=150, )

p.set_xlabel('x axis')
p.set_ylabel('y axis')
p.set_zlabel('z axis')


plt.show()

x.reshape(1,-1)
```
