```python
df2 = pd.DataFrame({'nama': ['one', 'one', 'one', 'two', 'two',
                            'two'],
                    'mobil': ['toyota', 'BMW', 'Ford', 'toyota', 'BMW', 'Ford'],
                    'sales': [1, 2, 3, 4, 5, 6]})
df2
```

```python
df2 = df2.pivot(index='nama', columns='mobil')
df2
```

```python
fig, ax = plt.subplots()
plt.imshow(df2, cmap="summer_r")
plt.colorbar()
# print(df2.columns[1])
col = list(map(lambda x:x[1], df2.columns.tolist()))
print(col)
row = list(map(lambda x:x, df2.index.tolist()))
print(row)


for i in range(len(row)):
    for j in range(len(col)):
#         print(df2.iloc[i,j])
        ax.text(j,i, df2.iloc[i,j], color="k", 
                ha="center",
                va='center',
                fontdict={'size':20, 'family':'arial'})

ax.set_xticks(np.arange(len(col)))
ax.set_xticklabels(col)
ax.set_yticks(np.arange(len(row)))
ax.set_yticklabels(row)
ax.set_ylim(-0.5,1.5)

plt.show()
```
