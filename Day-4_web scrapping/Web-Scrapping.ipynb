## Scrapping dari web Ultraman

```python
url = "http://www.scifijapan.com/articles/2015/10/04/bandai-ultraman-ultra-500-figure-list/"

dataultra = requests.get(url).content
# print(dataultra)
data4 = bs(dataultra,'html.parser')
nama = []
namaUltraman = []
namaMonster = []

for i in data4.find_all('strong'):
    nama.append(i.text)
    
namaUltraman = nama[2:36]
namaMonster = nama[37:110]
print(len(namaUltraman))
print(len(namaMonster))
print(namaMonster)

for i in namaMonster:
    print(i)
```

```python
url = "http://digidb.io/digimon-list/"

dataDigimon = requests.get(url).content
# print(dataultra)
dataDigi = bs(dataDigimon,'html.parser')

# print(dataDigi)
counter = 0
dataDigitoCSV = []
for i in dataDigi.find_all('td'):
#     print(counter)
    cek = str(i.text)
        
    print(cek, end = ' ')
    if cek[0]== ' ': print('y')
        
    temList.append(cek.replace('\xa0', ''))
#     temList.append(cek)
    counter += 1
    if (counter%13==0):
        dataDigitoCSV.append(temList)
        temList =[]
        print()
        counter = 0
# print(row)

for i in dataDigitoCSV:
    print(i)

```

```python
#scrapping src text from img tag 
image = []

for i in dataDigi.find_all('td'):
    if i.img:
        i.img['src']
#         print(i.img['src'])
        image.append(i.img['src'])
print("banyak image : ", len(image))
    
header = []
for i in dataDigi.find_all('th'):
    print(i.text)
    header.append(i.text)
header.append('image')
print(header)
print("banyak kolom : ", len(header))
```
